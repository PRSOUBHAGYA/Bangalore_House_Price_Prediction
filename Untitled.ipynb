```python
def median_are(sqft_range):
    temp = sqft_range.split('-')
    if len(temp) == 2:
        return (float(temp[0]) + float(temp[1]))/2
    try:
        return float(sqft_range)
    except:
        return None
```

```python
print(median_are('620 - 933'))
```

```python
df['total_sqft'] = df['total_sqft'].apply(median_are)
```

```python
def remove_outlier(df):
    final_df = pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m = np.mean(subdf.Price_per_Square_ft)
        sd = np.std(subdf.Price_per_Square_ft)
        gen_df = subdf[(subdf['Price_per_Square_ft'] > m-sd) &  (subdf['Price_per_Square_ft'] <= m+sd)]
#         print(subdf)
#         print("*"*20)
#         print(gen_df)
        
        final_df = pd.concat([final_df,gen_df],ignore_index=True)
    return final_df
        
df = remove_outlier(df)
```

```python
def remove_bhk_outlier(df):
    exclude_indices = np.array([])
    for key,subdf in df.groupby('location'):
        bhk_status = {}
        for bhk,bhk_df in df.groupby('bhk'):
            bhk_status[bhk]={
                'mean_bhk': np.mean(bhk_df['Price_per_Square_ft']),
                'count_bhk': bhk_df.shape[0]
            }
        for bhk,bhk_df in df.groupby('bhk'):
            stats = bhk_status.get(bhk-1)
            if stats and stats['count_bhk']>5:
                exclude_indices = np.append(exclude_indices,bhk_df[bhk_df['Price_per_Square_ft']<(stats['mean_bhk'])].index.values)
    return df.drop(exclude_indices,axis='index')
```

```python
df= remove_bhk_outlier(df)
```
